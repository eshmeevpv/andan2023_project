```python
import pandas as pd
```

```python
df_sales = pd.read_csv('vgsales.csv', sep=',')
df_sales.head()
```

```python
df_sales.drop({'Platform', 'Rank'}, axis= 1 , inplace= True)
df_sales.head()
```

```python
df_sales = df_sales.groupby(['Name', 'Year']).sum()
df_sales
```

```python
Twich_sort = pd.read_csv('Twich_sort.csv', sep=',')
Twich_sort.rename(columns = {'Game':'Name'}, inplace = True )
Twich_sort
```

```python
both = df_sales.merge(Twich_sort, on=["Name"])
both
```

```python
both.to_csv('Both.csv', sep=',')
```
